```python
import pandas as pd
from datetime import datetime

print ("MEETING MINUTES GENERATOR\n")

# User Inputs

meeting_title = input("Enter meeting title: ")

date = input ("Enter meeting date (DD-MM-YYYY ")

time = input("Enter meeting time: ")

participants = input("Enter participants (comma separated): ")

agenda = input("Enter agenda points (comma separated): ")

decisions = input("Enter decisions taken (comma separated): ")

action_items = input("Enter action items (comma separated): ") # Corrected variable name from sction_items

prepared_by = input("Prepared by: ") # Corrected prompt


participants_list = [p.strip() for p in participants.split(",") if p.strip()]
agenda_list = [a.strip() for a in agenda.split(",") if a.strip()] # Added . before split
decisions_list = [d.strip() for d in decisions.split(",") if d.strip()] # Added . before split and fixed variable name
actions_list = [a.strip() for a in action_items.split(",") if a.strip()]

# Create structured meeting minutes
minutes = { # Changed variable name to 'minutes' to avoid conflict with the type 'Minutes'

"Meeting Title": meeting_title,

"Date": date,

"Time": time,

"Participants": ", ".join(participants_list),

"Agenda": "; ".join(agenda_list),

"Decisions Taken": "; ".join(decisions_list),

"Action Items": "; ".join(actions_list),

"Prepared By": prepared_by,

"Generated On": datetime.now().strftime("%d-%m-%Y %H:%M:%S") # Fixed %Xm to %M
}


# Convert to DataFrame

df = pd.DataFrame(list(minutes.items()), columns=["Field", "Details"])

print("\n Meeting Minutes Generated Successfully\n")

display(df)


df.to_csv("meeting_minutes.csv", index=False)

print("\n File saved as meeting_minutes.csv")
```
